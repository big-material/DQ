```python
from pathlib import Path
import os

# create dir processed_data if it doesn't exist
processed_data_dir = Path("processed_data")
if not processed_data_dir.exists():
    processed_data_dir.mkdir()
```

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv('datasets/Bala_classification_dataset.csv')
df = df.drop(labels=["Me3"], axis=1)
df_category = df.select_dtypes(exclude=[np.number])

le = LabelEncoder()
for col in df_category.columns:
    df[col] = le.fit_transform(df[col])
df.to_csv('processed_data/Bala_classification_dataset.csv', index=False)
df
```
